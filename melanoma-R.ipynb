### Regression Analysis and correlation computation for the following given below

* time ~ thickneess

#### 1. Regression Analysis

y=b0 +b1*x + e
 where :
 * b0 is the intercept
 * b1, b2,b3,...,bn are the regression weights or coefficients
 * e is the error term

```r
# Regression Analysis for time ~ thickness
# attach(mel)
# lm(time ~ thickness)
# summary()
model <- lm(time ~ thickness, data = mel)
summary(model)$coef
```

The output above shows the estimate of the regression beta coefficients (column Estimate) and their significance levels (column Pr(>|t|). The intercept (b0) is 2413.41019 and the coefficient of thickness variable is -89.25454.

The estimated regression equation can be written as follow: time = 2413.41019 + -89.25454* thickness. Using this formula, for each new tumour thickness, you can predict the survival time.

For a new tumour thickness of say,10mm, we can predict the survival time to be 2413.41019 + -89.25454*10 ~ 1520 days ~4 months.

The code below agrees with our analysis

```r
newdata <- data.frame(thickness= c(0,  10))
model %>% predict(newdata)
```

```r
# Regression Analysis for time ~ age
model <- lm(time ~ age, data = mel)
summary(model)$coef
```

The output above shows the estimate of the regression beta coefficients (column Estimate) and their significance levels (column Pr(>|t|). The intercept (b0) is 3217.44757 and the coefficient of age variable is -20.29314.

The estimated regression equation can be written as follow: time = 3217.44757 + -20.29314* age. Using this formula, for each new age of patients, you can predict the survival time.

For a new patients of age say 40 years we can predict the survival time to be 3217.44757 + -20.29314*40 ~ 2405days ~6.5months.

The code below agrees with our analysis

```r
newdata <- data.frame(age= c(0,  40))
model %>% predict(newdata)
```

```r
#correlation computation for time ~ thickness
attach(mel)
corr=cor.test(time,thickness,method='pearson')
corr
```

Since the Pearson coefficient is -0.2354087. Also p-value is 0.0006793 < 0.05. So the null hypothesis is rejected. So it is statistically reasonable to conclude that there is significant negative correlation between the survival time and thickness based on the sample.

### Two Sample significance test

* Two sample  Z-test (Large Sample) 

First, we group the data by gender by creating two datasets. One for male and the other for female categories

```r
mel_female <- mel[mel$sex==0,]
head(mel_female)
```

```r
mel_male <- mel[mel$sex==1,]
head(mel_male)
```

### Two sample z test between the variables  for the Female gender

```r
library(BSDA)
```

```r
# Two sample z test for time ~ thickness for female gender
z.test(x=mel_female$time, y=mel_female$thickness, mu=0, sigma.x=15, sigma.y=15)
```

The test statistic for the two sample z-test is 1206.5 and the corresponding p-value is 2.2e-16. Since this p-value is less than .05, we  have sufficient evidence to accepts alternative hypothesis. Thus, we conclude that the mean level is significantly different between the two variables (time~thickness)

```r
# Two sample z test between time~age for female gender
z.test(x=mel_female$time, y=mel_female$age, mu=0, sigma.x=15, sigma.y=15)
```

The test statistic for the two sample z-test is 1180.6 and the corresponding p-value is 2.2e-16. Since this p-value is less than .05, we  have sufficient evidence to accepts alternative hypothesis. Thus, we conclude that the mean level is significantly different between the two variables (time~age)

```r
## Two sample z test between thickness and age for females
z.test(x=mel_female$thickness, y=mel_female$age, mu=0, sigma.x=15, sigma.y=15)
```

The test statistic for the two sample z-test is -25.969 and the corresponding p-value is 2.2e-16. Since this p-value is less than .05, we  have sufficient evidence to accepts alternative hypothesis. Thus, we conclude that the mean level is significantly different between the two variables (thickness~age)

diagnostic plots show residuals in four different ways:

Residuals vs Fitted. Used to check the linear relationship assumptions. A horizontal line, without distinct patterns is an indication for a linear relationship, what is good. This is not the case in our example.

Normal Q-Q. Used to examine whether the residuals are normally distributed. It’s good if residuals points follow the straight dashed line. This is not the case in our example.

Scale-Location (or Spread-Location). Used to check the homogeneity of variance of the residuals (homoscedasticity). Horizontal line with equally spread points is a good indication of homoscedasticity. This is not the case in our example, where we have a heteroscedasticity problem.

Residuals vs Leverage. Used to identify influential cases, that is extreme values that might influence the regression results when included or excluded from the analysis. The plot-4 above highlights the top 2 most extreme points (#43 and #21), with a standardized residuals above 0.5.
